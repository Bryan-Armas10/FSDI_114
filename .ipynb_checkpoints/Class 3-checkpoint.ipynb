```python
# Linked List

# In General,, there are two varietes of linked list, and these are described by how nodes in the list are connected to each other.

    # 1. Singly linked list (uni-directional)
    # 2. Doubly linked list (bi-directional)
```

```python
class SinglyLinkedList:
    class __Node:
        def __init__(self, data):
            self.data = data
            self.next = None

    def __init__(self):
        self.head = None
        self.tail = None
        self.count = 0

    def append(self, datum):
        new_node = self.__Node(datum)
        self.count += 1
        if not self.head:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node

    def insert(self, index, datum):
        # this method inserts new node before the target index.
        # However: if the target index is not present in the list, then this is equivalent to append.
        # Note: It is possible to replace the head or tail of the list with this method
        # Note 2: Also, if the list is empty,, then again, this function the same as append 

        new_node = self.__Node(datum)

        # Case 1: The list is empty, just insert the node.
        if self.head is None:
            self.head = new_node
            self.tail = new_node
            self.count += 1
            return
    
        # Case 2: Insert at the beginning (if index is 0 or if the list is empty)
        if index == 0:
            new_node.next = self.head
            self.head = new_node
            self.count += 1
            return
    
        # Case 3: Insert at a specific position (middle or end)
        current = self.head
        prev = None
        current_index = 0
    
        # Traverse the list to find the target index
        while current:
            if current_index == index:
                # Insert the new node before the target node
                prev.next = new_node
                new_node.next = current
                if current == self.tail:
                    self.tail = new_node
                self.count += 1
                return
            prev = current
            current = current.next
            current_index += 1

        # Case 4: If the index is not found, append the node at the end
        prev.next = new_node
        self.tail = new_node
        self.count += 1

    def index(self, datum):             #  {1, 2, 3}
        current = self.head
        found = False 
        count = 0
        while current:
            if current.data == datum:
                found = True
                break
            current = current.next
            count += 1
        if found:
            return count
        raise("Value not found")

    def remove(self, datum):
        current = self.head
        prev = None
        found = False
        while current:
            if current.data == datum:
                found = True
                self.count -= 1                 # errata
                break
            prev = current
            current = current.next
        if found:
            if prev:
                prev.next = current.next
                if current == self.tail and prev.next:
                    self.tail = prev
            else:
                self.head = self.head.next
                if not self.head:
                    self.tail = None
        else:
            raise ValueError("Value not found")

    def __len__(self):
        return count

    def __str__(self):
        current = self.head
        out = "["
        if current:
            out += "%s" % repr(current.data)
            current = current.next
            while current:                                # while current is not equal to None 
                out += ", %s" % repr(current.data)
                current = current.next
        out += "]"
        return out
```

```python
# pylist = []
sll = SinglyLinkedList()

print(sll)

sll.append(1)
print(sll)

sll.append(2)
print(sll)

sll.append(3)
print(sll)

# remove
print("Remove example:")
sll.remove(3)
print(sll)

sll.append(4)
print(sll)
```
